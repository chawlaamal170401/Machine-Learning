```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Line Plot

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
x = np.array([1, 2, 3])
y = np.array([2, 4, 1])
plt.plot(x, y, marker="o", color="red", markerfacecolor="yellow")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot")
plt.show()
```

```python
x = np.arange(6)
y = x**2
z = 2*x + 3
plt.figure(figsize=(10, 5))
plt.style.use('seaborn')
plt.plot(x, y, marker="*", color="green", markerfacecolor="orange", label="Apple", markersize=12)
plt.plot(x, z, marker="^", color="red", markerfacecolor="blue", label="Kiwi", linestyle="dashed", linewidth=5, markersize=10)
plt.title("Price of Fruits over time")
plt.xlabel("Time")
plt.ylabel("Fruits Price")
plt.legend()
plt.show()
```

### Bar Chart

```python
x = np.arange(3)
y = [10, 20, 30]
z = [30, 10, 15]

tick_label = ['Apple', 'Grapes', 'Kiwi']
plt.bar(x-0.25, y, tick_label=tick_label, width=0.25, label="2020", color="red")
plt.bar(x, z, label="2021", width = 0.25, color="pink")
plt.title("Prices of Fruits Over Time")
plt.xlabel("Time")
plt.ylim(0,40)
plt.ylabel("Prices")
plt.legend()
plt.show()
```

### Pie Chart

```python
subjects = ['Maths', 'Physics', 'Chemistry', 'ComputerScience']
marks = [95, 94, 90, 89]
colors = ['r', 'y', 'orange', 'pink']
plt.pie(marks, labels=subjects, colors=colors, shadow=True, explode=(0, 0, 0, 0.2), autopct='%1.1f%%')
plt.title("Marks In Different Subjects")
plt.show()
```

### Scatter Plots

```python
x = np.arange(10)
y = np.array([2,4,5,7,6,8,9,11,12,12])
plt.scatter(x, y, color="red", marker="*")
plt.title("Scatter Plot")
plt.show()
```

### Histogram

```python
# Standard Normal Deviation
Xsn = np.random.randn(100)
sigma = 8
u = 70
X = np.round(Xsn*sigma + u)
X1 = np.round(Xsn*5 + 40)
print(X, X1)
```

```python
plt.hist(X, alpha=0.8)
plt.hist(X1, alpha=1)
plt.ylabel("Probability/Frequency of Students")
plt.xlabel("Marks Range")
plt.show()
```

## Matplotlib With Pandas

```python
df = pd.read_csv("./company_sales_data.csv")
df.head(10)
```

### Q1. Read Total profit of all months and show it using a line plot

```python
x = np.array(list(df['month_number']))
y = np.array(list(df['total_profit']))
plt.style.use("seaborn")
plt.plot(x, y, linewidth=5, color="red")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Company Profit per Month")
plt.show()
```

### Q2. Get total profit of all months and show line plot with the following Style properties
#### Generated line plot must include following Style properties: –

- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3


```python
x = np.array(list(df['month_number']))
y = np.array(list(df['total_profit']))
plt.style.use("classic")
plt.plot(x, y, color="red", linestyle="dashed", linewidth=3, marker='o', markerfacecolor='k')
plt.xlabel("Month Number")
plt.ylabel("Sold Units Number")
plt.xticks(x)
plt.xlim(0, 15)
plt.title("Company Sales Data of Last Year")
plt.show()
```

### Q3. Read all product sales data and show it  using a multiline plot

```python
monthList  = np.array(list(df ['month_number']))
faceCremSalesData   = np.array(list(df ['facecream']))
faceWashSalesData   = np.array(list(df ['facewash']))
toothPasteSalesData = np.array(list(df ['toothpaste']))
bathingsoapSalesData   = np.array(list(df ['bathingsoap']))
shampooSalesData   = np.array(list(df ['shampoo']))
moisturizerSalesData = np.array(list(df ['moisturizer']))
plt.style.use("seaborn")

plt.plot(monthList, faceCremSalesData,   label = 'Face cream Sales Data', marker='o', linewidth=3)
plt.plot(monthList, faceWashSalesData,   label = 'Face Wash Sales Data',  marker='*', linewidth=3)
plt.plot(monthList, toothPasteSalesData, label = 'ToothPaste Sales Data', marker='^', linewidth=3)
plt.plot(monthList, bathingsoapSalesData, label = 'ToothPaste Sales Data', marker='X', linewidth=3)
plt.plot(monthList, shampooSalesData, label = 'ToothPaste Sales Data', marker='P', linewidth=3)
plt.plot(monthList, moisturizerSalesData, label = 'ToothPaste Sales Data', marker='D', linewidth=3)

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(monthList)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
```

### Q4. Read toothpaste sales data of each month and show it using a scatter plot

```python
toothPasteSalesData = np.array(list(df ['toothpaste']))
plt.scatter(monthList, toothPasteSalesData, label = 'Tooth paste Sales data', color='red')
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.legend(loc='upper left')
plt.title(' Tooth paste Sales data')
plt.xticks(monthList)
plt.grid(True, linewidth= 1, linestyle="--")
plt.show()
```

### Q5. Read face cream and facewash product sales data and show it using the bar chart


```python
plt.bar([x-0.25 for x in monthList], faceCremSalesData, width= 0.25, label = 'Face Cream sales data', align='edge')
plt.bar([x+0.25 for x in monthList], faceWashSalesData, width= -0.25, label = 'Face Wash sales data', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')

plt.xticks(monthList)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Facewash and facecream sales data')
plt.show()
```

### Q6. Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

```python
plt.bar(monthList, bathingsoapSalesData, color="green")
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title(' Sales data')
plt.xticks(monthList)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('bathingsoap sales data')
plt.savefig('D:\Data Science\Matplotlib', dpi=150)
plt.show()
```

### Q7. Read the total profit of each month and show it using the histogram to see the most common profit ranges


```python
profitList = np.array(list(df['total_profit']))
labels = ['low', 'average', 'Good', 'Best']
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profitList, profit_range, label = 'Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.xticks(profit_range)
plt.title('Profit data')
plt.ylim(0, 8)
plt.show()
```

### Q8. Calculate total sale data for last year for each product and show it using a Pie chart

```python
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), 
         df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%1.1f%%')
plt.legend(loc='upper left')
plt.title('Sales data')
plt.show()
```

### Q9. Read Bathing soap facewash of all months and display it using the Subplot


```python
bathingsoap = np.array(list(df['bathingsoap']))
f, arr = plt.subplots(2, sharex=True)
arr[0].plot(monthList, bathingsoap, label = 'Bathingsoap Sales Data', color='blue', marker='o', linewidth=3)
arr[0].set_title('Sales data of  a Bathingsoap')
arr[1].plot(monthList, faceWashSalesData, label = 'Face Wash Sales Data', color='black', marker='^', linewidth=3)
arr[1].set_title('Sales data of  a facewash')

plt.xticks(monthList)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

### Q10. Read all product sales data and show it using the stack plot

```python
plt.plot([],[],color='m', label='face Cream', linewidth=5)
plt.plot([],[],color='c', label='Face wash', linewidth=5)
plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)

plt.stackplot(monthList, faceCremSalesData, faceWashSalesData, toothPasteSalesData, 
              bathingsoapSalesData, shampooSalesData, moisturizerSalesData, 
              colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('All product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
