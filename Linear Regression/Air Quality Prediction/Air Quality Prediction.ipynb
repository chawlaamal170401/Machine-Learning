# Linear Regression

```python
def hypothesis(x, theta):
    return np.dot(x, theta)
```

```python
def error(x, y, theta):
    e = 0.0
    N = x.shape[0]
    y_ = hypothesis(x, theta)
    e = np.sum((y-y_)**2)
    
    return e/N
```

```python
def gradient(x, y, theta):
    N = x.shape[0]
    y_ = hypothesis(x, theta)
    grad = np.dot(x.T, y_-y)
    
    return grad/N
```

```python
def gradient_descent(x, y, learning_rate=0.01, max_iterations=500):
    error_list = []
    n = x.shape[1]
    theta = np.zeros((n, ))
    
    for i in range(max_iterations):
        e = error(x, y, theta)
        error_list.append(e)
        
        grad = gradient(x, y, theta)
        theta = theta - learning_rate*grad
        
    return theta, error_list
```

```python
theta, error_list = gradient_descent(x_train, y_train)
```

```python
theta
```

```python
plt.plot(error_list)
plt.show()
```

```python
def score(y, y_):
    num = np.sum((y-y_)**2)
    den = np.sum((y-y.mean())**2)
    score = 1-(num/den)
    
    return score*100
```

```python
y_ = hypothesis(x_train, theta)
train_score = score(y_train, y_)
train_score
```
