```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression
```

```python
data = np.genfromtxt('train.csv', delimiter=',')
data
```

```python
type(data)
print(data.shape)
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
x_train = df.iloc[:, :10]
x_train.head()
```

```python
y_train = df.iloc[:, 10]
y_train.head()
```

```python
x_train.isnull().sum()
```

```python
alg = LinearRegression()
```

```python
test_data = np.genfromtxt('test.csv', delimiter=',')
test_data
df_test = pd.DataFrame(test_data)
```

```python
df_test.head()
df_test
```

```python
x_test = df_test.iloc[:, :10]
```

```python
x_test
```

```python
alg.fit(x_train, y_train)
```

```python
y_pred = alg.predict(x_test)
```

```python
m = alg.coef_
```

```python
c = alg.intercept_
```

```python
train_score = alg.score(x_train, y_train)
```

```python
print(train_score)
```

```python
test_score = alg.score(x_test, y_pred)
print(test_score)
```
